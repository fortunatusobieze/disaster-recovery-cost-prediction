```python
import requests
import pandas as pd
import numpy as np

# Base URL for all FEMA API calls
BASE_URL = "https://www.fema.gov/api/open"

print("Libraries loaded successfully")
print(f"Pandas version: {pd.__version__}")
print(f"Numpy version: {np.__version__}")
```

```python
# Fetch Dataset 1 - Disaster Declarations Summaries

url = f"{BASE_URL}/v2/DisasterDeclarationsSummaries"

params = {
    "$top": 200,       # Fetch only 200 records for now (exploration)
    "$format": "json"  # Return data as JSON
}

# Make the API request
response = requests.get(url, params=params)

# Check if request was successful
print(f"Status Code: {response.status_code}")  # 200 = success

# Parse the JSON response
data = response.json()

# Convert to DataFrame
df_declarations = pd.DataFrame(data["DisasterDeclarationsSummaries"])

# Quick overview
print(f"Shape: {df_declarations.shape}")
print(f"\nColumn Names:")
print(df_declarations.columns.tolist())
print(f"\nFirst 3 rows:")
display(df_declarations.head(3))
```

```python
# Fetch Dataset 3 - FEMA Web Disaster Summaries

print("=" * 55)
print("DATASET 3: FEMA WEB DISASTER SUMMARIES")
print("=" * 55)

url3 = f"{BASE_URL}/v1/FemaWebDisasterSummaries"

params3 = {
    "$top": 200,
    "$format": "json"
}

response3 = requests.get(url3, params=params3)
print(f"Status Code: {response3.status_code}")

data3 = response3.json()

# Check the exact key name in the response
print(f"\nResponse keys: {list(data3.keys())}")

# Get the correct key
key3 = [k for k in data3.keys() if k != "metadata"][0]
df_summaries = pd.DataFrame(data3[key3])

print(f"Shape: {df_summaries.shape}")

# All column names
print(f"\nAll Columns:")
for col in df_summaries.columns.tolist():
    print(f"  - {col}")

# Null counts
print("\n── NULL COUNTS ──")
null_counts3 = df_summaries.isnull().sum()
print(null_counts3[null_counts3 > 0] if null_counts3.sum() > 0 else "No nulls found")

# Key columns preview
print("\n── FIRST 5 ROWS ──")
display(df_summaries.head(5))
```

```python
# Verifying Join Key Across All 3 Datasets

print("=" * 55)
print("JOIN KEY VERIFICATION: disasterNumber")
print("=" * 55)

# Confirm disasterNumber exists in all 3
print("\n── COLUMN EXISTS CHECK ──")
print(f"Dataset 1 (Declarations):  {'disasterNumber' in df_declarations.columns} ✅")
print(f"Dataset 2 (PA Projects):   {'disasterNumber' in df_pa.columns} ✅")
print(f"Dataset 3 (Summaries):     {'disasterNumber' in df_summaries.columns} ✅")

# Check data types match
print("\n── DATA TYPE CHECK ──")
print(f"Dataset 1 type: {df_declarations['disasterNumber'].dtype}")
print(f"Dataset 2 type: {df_pa['disasterNumber'].dtype}")
print(f"Dataset 3 type: {df_summaries['disasterNumber'].dtype}")

# Convert all to same type for safe joining
df_declarations['disasterNumber'] = df_declarations['disasterNumber'].astype(str)
df_pa['disasterNumber'] = df_pa['disasterNumber'].astype(str)
df_summaries['disasterNumber'] = df_summaries['disasterNumber'].astype(str)

print("\n── AFTER STANDARDISATION (all converted to string) ──")
print(f"Dataset 1 type: {df_declarations['disasterNumber'].dtype}")
print(f"Dataset 2 type: {df_pa['disasterNumber'].dtype}")
print(f"Dataset 3 type: {df_summaries['disasterNumber'].dtype}")

# Check overlapping disaster numbers
common_1_2 = set(df_declarations['disasterNumber']) & set(df_pa['disasterNumber'])
common_1_3 = set(df_declarations['disasterNumber']) & set(df_summaries['disasterNumber'])
common_all = set(df_declarations['disasterNumber']) & \
             set(df_pa['disasterNumber']) & \
             set(df_summaries['disasterNumber'])

print(f"\n── OVERLAP CHECK ──")
print(f"Disasters in BOTH Dataset 1 & 2:       {len(common_1_2)}")
print(f"Disasters in BOTH Dataset 1 & 3:       {len(common_1_3)}")
print(f"Disasters in ALL 3 datasets:           {len(common_all)}")

# Sample values from each
print(f"\n── SAMPLE DISASTER NUMBERS ──")
print(f"Dataset 1: {df_declarations['disasterNumber'].head(5).tolist()}")
print(f"Dataset 2: {df_pa['disasterNumber'].head(5).tolist()}")
print(f"Dataset 3: {df_summaries['disasterNumber'].head(5).tolist()}")
```

```python
# Prove the Join Works on Overlapping Records

print("=" * 55)
print("PROVING THE JOIN WORKS")
print("=" * 55)

# Join Dataset 1 & 3 on disasterNumber (14 overlaps exist)
df_joined = pd.merge(
    df_declarations,
    df_summaries,
    on='disasterNumber',
    how='inner'  # Only keep rows that exist in BOTH tables
)

print(f"\nDataset 1 rows:          {len(df_declarations)}")
print(f"Dataset 3 rows:          {len(df_summaries)}")
print(f"Joined result rows:      {len(df_joined)}")
print(f"\n✅ Join works! {len(df_joined)} matching disasters found")

# Show the joined result
print("\n── JOINED DATA PREVIEW ──")
preview_cols = ['disasterNumber', 'state', 'incidentType',
                'declarationDate', 'totalObligatedAmountPa']
display(df_joined[preview_cols].head(10))
```
